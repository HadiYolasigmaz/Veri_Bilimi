```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x değişkeni - 0 ile 10 arasını 500 eşit parçaya bölelim np.linspace
x = np.linspace(0, 10, 500)
# ye değişkeni bu 500 parçanın her birinin sinusu
y = np.sin(x)

fig, ax = plt.subplots()

# dashes() metodunu kullanma
line1, = ax.plot(x, y, label='dashes() metodunu kullanma')
 
# 2pt çizgi, 2pt boş, 10pt çizgi, 2pt boş
line1.set_dashes([2, 2, 10, 2]) 

#plot(..., dashes=...) metodu ile bir çizgi oluşturalım
line2, = ax.plot(x, y - 0.2, dashes=[6, 2], label='Tire parametresini kullanma')

ax.legend()
plt.show()
```

```python
# aynı sayıları üretmek istiyoruz
np.random.seed(19680801)

# x y ve z değişkenlerini tanımlıyoruz
x = np.random.rand(10)
y = np.random.rand(10)
z = np.sqrt(x**2 + y**2)

plt.subplot(321)
plt.scatter(x, y, s=80, c=z, marker=">")

plt.subplot(322)
plt.scatter(x, y, s=80, c=z, marker=(5, 0))

verts = np.array([[-1, -1], [1, -1], [1, 1], [-1, -1]])
plt.subplot(323)
plt.scatter(x, y, s=80, c=z, marker=verts)

plt.subplot(324)
plt.scatter(x, y, s=80, c=z, marker=(5, 1))

plt.subplot(325)
plt.scatter(x, y, s=80, c=z, marker='+')

plt.subplot(326)
plt.scatter(x, y, s=80, c=z, marker=(5, 2))

plt.show()
```

```python
# aynı sayıları üretmek istiyoruz
np.random.seed(19680801)


x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
s = np.random.rand(*x.shape) * 800 + 500

plt.scatter(x, y, s, c="g", alpha=0.5, marker=r'$\clubsuit$',
            label="Şans")
plt.xlabel("Dağılım")
plt.ylabel("Altın")
plt.legend(loc='upper left')
plt.show()
```

```python
# Aynı rastgele sayıları üretmek istiyoruz
np.random.seed(19680801)
dt = 0.001
t = np.arange(0.0, 10.0, dt)
#dürtü yanıtı
r = np.exp(-t[:1000] / 0.05) 
x = np.random.randn(len(t))
# gürültü renklendirme
s = np.convolve(x, r)[:len(x)] * dt  

fig, ana_ax = plt.subplots()
ana_ax.plot(t, s)
ana_ax.set_xlim(0, 1)
ana_ax.set_ylim(1.1 * np.min(s), 2 * np.max(s))
ana_ax.set_xlabel('zaman (s)')
ana_ax.set_ylabel('akım (nA)')
ana_ax.set_title('Gauss renkli gürültü')

# Olasılık grafiği
right_inset_ax = fig.add_axes([.65, .6, .2, .2], facecolor='r')
right_inset_ax.hist(s, 400, density=True)
right_inset_ax.set_title('Olasılık')
right_inset_ax.set_xticks([])
right_inset_ax.set_yticks([])

# dürtü yanıtı grafiği
left_inset_ax = fig.add_axes([.2, .6, .2, .2], facecolor='r')
left_inset_ax.plot(t[:len(r)], r)
left_inset_ax.set_title('dürtü yanıtı')
left_inset_ax.set_xlim(0, 0.2)
left_inset_ax.set_xticks([])
left_inset_ax.set_yticks([])

plt.show()
```
