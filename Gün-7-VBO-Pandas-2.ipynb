```python
## artık elimizde 2 kütüphane var olacak numpy ve pandas
import numpy as np 
import pandas as pd
```

```python
## pandada dataframe kavramına anlamak için rasgele sayı üretmeyi planlıyorum.
## bu yüzden numpy kütüphanesinden sadece numpy random metodunun içinde olan randn çağırıyorum
from numpy.random import randn
## ve bildiğiniz gibi her seferinde aynı sayıyı üretmesi içinde np.random.seed i kullanacağım.
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(5,4), index='Ankara İstanbul İzmir Bursa Antalya'.split(),
                 columns = 'Otobüs Uçak Taksi Otomobil'.split())
```

```python
df
```

```python
## Otobüs,taksi ve otomobilden oluşan kara araçlarındaki değeri toplayıp 
## yeni bir kolon ekleyerek adına karadan adını vermek ister isek.
df['Karadan'] = df['Otobüs'] + df['Taksi'] + df['Otomobil']
```
