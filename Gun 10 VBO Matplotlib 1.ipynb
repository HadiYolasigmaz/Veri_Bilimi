```python
import numpy as np

import matplotlib.pyplot as plt

# 5 Sınav sonucu Matematik edebiyat
N = 5
matematik_ortalamaları = (30, 35, 30, 35, 27)
# sınav sırası
endeks = np.arange(N)   
bar_genisligi = 0.35       # the width of the bars: can also be len(x) sequence
#liste1
plt.bar(endeks, matematik_ortalamaları, bar_genisligi)

plt.ylabel('Notlar')
plt.title('Matematik 5 sınav sonucları')
plt.xticks(endeks, ('Sınav1', 'Sınav2', 'Sınav3', 'Sınav4', 'Sınav5'))
plt.yticks(np.arange(0, 41, 10))


plt.show()
```

```python
# 5 Sınav sonucu Matematik edebiyat
N = 5
matematik_ortalamaları = (20, 35, 30, 35, 27)
edebiyat_ortalamaları = (25, 32, 34, 20, 25)
# sınav sırası
endeks = np.arange(N)   
bar_genisligi = 0.55       # the width of the bars: can also be len(x) sequence
#liste1
p1 = plt.bar(endeks, matematik_ortalamaları, bar_genisligi)
#liste1
p2 = plt.bar(endeks, edebiyat_ortalamaları, bar_genisligi,bottom=matematik_ortalamaları)

plt.ylabel('Notlar')
plt.title('Matematik ve edebiyat 5 sınav sonucları')
plt.xticks(endeks, ('Sınav1', 'Sınav2', 'Sınav3', 'Sınav4', 'Sınav5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Matematik', 'Edebiyat'))

plt.show()
```

```python
# 5 Sınav sonucu Matematik edebiyat
N = 5
matematik_ortalamaları = (20, 35, 30, 35, 27)
edebiyat_ortalamaları = (25, 32, 34, 20, 25)
matematik_standartsapma = (2, 3, 4, 1, 2)
edebiyat_standartsapma = (3, 5, 2, 3, 3)
# sınav sırası
endeks = np.arange(N)   
bar_genisligi = 0.35       

p1 = plt.bar(endeks, matematik_ortalamaları, bar_genisligi, yerr=matematik_standartsapma)
#liste1
p2 = plt.bar(endeks, edebiyat_ortalamaları, bar_genisligi,bottom=matematik_ortalamaları, yerr=edebiyat_standartsapma)

plt.ylabel('Notlar')
plt.title('Matematik ve edebiyat 5 sınav sonucları')
plt.xticks(endeks, ('Sınav1', 'Sınav2', 'Sınav3', 'Sınav4', 'Sınav5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Matematik', 'Edebiyat'))

plt.show()
```

```python
data = {'elma': 10, 'armut': 15, 'portakal': 5, 'muz': 20}

degiskenler = list(data.keys())

degerler = list(data.values())

# 3 tane grafikten oluşacak diyoruz
fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)

# bar grafiğini çiz
axs[0].bar(degiskenler, degerler)  

# bar grafiğini çiz
axs[1].scatter(degiskenler, degerler)

# dağılım grafiğini çiz
axs[2].plot(degiskenler, degerler)

# çizgi grafiğini çiz
fig.suptitle('Kategorilere göre seçim')
```

```python
# Örnek Veri oluşturalım - Rastgele
x = np.arange(0.1, 4, 0.1)
y = np.exp(-x)

# Örnek hata değişkenleri hata barı
ozel_fonksiyon=np.sqrt(x)

fig, axs = plt.subplots(nrows=1, ncols=2, sharex=True)
ax = axs[0]
ax.errorbar(x, y, yerr=ozel_fonksiyon)
#ax.errorbar(x, y, yerr=0.3)
ax.set_title('Tüm hatalar')

ax = axs[1]
ax.errorbar(x, y, yerr=ozel_fonksiyon, errorevery=5)
ax.set_title('sadece 5. hatalar')


fig.suptitle('Hataları gösterirken hangi grafiği kullanalım')

plt.show()
```
