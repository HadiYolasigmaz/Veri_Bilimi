```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# capı 3 olan daire çizelim.

an = np.linspace(0, 2 * np.pi, 100)
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[0, 0].set_title('elips', fontsize=10)

axs[0, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[0, 1].axis('equal')
axs[0, 1].set_title('daire', fontsize=10)

axs[1, 0].plot(3 * np.cos(an), 3 * np.sin(an))
axs[1, 0].axis('equal')
axs[1, 0].set(xlim=(-3, 3), ylim=(-3, 3))
axs[1, 0].set_title('limitleri değiştirsek bile daire', fontsize=10)

axs[1, 1].plot(3 * np.cos(an), 3 * np.sin(an))
axs[1, 1].set_aspect('equal', 'box')
axs[1, 1].set_title('veri aralığı otomatik değiştirdik gene daire', fontsize=10)

fig.tight_layout()

plt.show()
```

```python
# Aynı rastgele sayıları üretmek istiyoruz
np.random.seed(19680801)
 
N_points = 100000
n_bins = 50

# Merkezi x=0 ve y=5 olan bir normal dağılım oluşturmak istiyoruz.
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(100000) + 5

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# bins ile aralık sayısını tespit ediyoruz.
axs[0].hist(x, bins=n_bins)
axs[1].hist(y, bins=n_bins)
```

```python
# Aynı rastgele sayıları üretmek istiyoruz
np.random.seed(19680801)

mu = 200
sigma = 25
x = np.random.normal(mu, sigma, size=100)

fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(8, 4))

ax0.hist(x, 20, density=True, histtype='stepfilled', facecolor='g', alpha=0.75)
ax0.set_title('Doldurulmus')

bins = [100, 150, 180, 195, 205, 220, 250, 300]
ax1.hist(x, bins, density=True, histtype='bar', rwidth=0.8)
ax1.set_title('Doldurulmamış')

fig.tight_layout()
plt.show()
```

```python
# Aynı rastgele sayıları üretmek istiyoruz
np.random.seed(19680801)

n_bins = 10
x = np.random.randn(1000, 3)

fig, axes = plt.subplots(nrows=2, ncols=2)
ax0, ax1, ax2, ax3 = axes.flatten()

colors = ['red', 'tan', 'lime']
label_colors = ['kirmizi', 'kahverengi', 'yesil']
ax0.hist(x, n_bins, density=True, histtype='bar', color=colors, label=label_colors)
ax0.legend(prop={'size': 10})
ax0.set_title('Çubuk adları')

ax1.hist(x, n_bins, density=True, histtype='bar', stacked=True)
ax1.set_title('Çubuk Yığınları')

ax2.hist(x, n_bins, histtype='step', stacked=True, fill=False)
ax2.set_title('Yığın adımları Doldurulmamış')

# Make a multiple-histogram of data-sets with different length.
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax3.hist(x_multi, n_bins, histtype='bar')
ax3.set_title('Farklı örnek boyutları')

fig.tight_layout()
plt.show()
```

```python
# Pasta kartı, pasta sırasına göre
etiketler = 'Kedi', 'Köpek', 'Eşek', 'Katır'
boyutlar = [15, 30, 45, 10]
ayir = (0, 0.2, 0, 0)  

fig1, ax1 = plt.subplots()
ax1.pie(boyutlar, explode=ayir, labels=etiketler, autopct='%1.1f%%',
        shadow=True, startangle=90)
# Equal dairede orantısal dağılım oldugunu gösterir.
ax1.axis('equal')  

plt.show()
```

```python
# Verilerimiz
etiketler = 'Kedi', 'Köpek', 'Eşek', 'Katır'
oranlar = [15, 30, 45, 10]

# Make figure and axes
fig, axs = plt.subplots(2, 2)

# A standard pie plot
axs[0, 0].pie(oranlar, labels=etiketler, autopct='%1.1f%%', shadow=True)

# Shift the second slice using explode
axs[0, 1].pie(oranlar, labels=etiketler, autopct='%.0f%%', shadow=True, explode=(0, 0.3, 0, 0))

# Adapt radius and text size for a smaller pie
patches, texts, autotexts = axs[1, 0].pie(oranlar, labels=etiketler,
                                          autopct='%.0f%%', )
# Make percent texts even smaller
plt.setp(autotexts, size='x-small')
autotexts[0].set_color('white')
autotexts[1].set_color('white')

# Use a smaller explode and turn of the shadow for better visibility
patches, texts, autotexts = axs[1, 1].pie(oranlar, labels=etiketler,
                                          autopct='%.0f%%',
                                          shadow=False, radius=0.5,
                                          explode=(0, 0.05, 0, 0))
plt.setp(autotexts, size='x-small')
autotexts[0].set_color('white')

plt.show()
```

```python
# Aynı rastgele sayıları üretmek istiyoruz
np.random.seed(19680801)

# pasta dilimlerini hesapla
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

ax = plt.subplot(111, projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5)

plt.show()
```

```python
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("polar axiste bir doğru çizimi", va='bottom')
plt.show()
```
